#### 1. How To Find The Most Popular Movies

For this notebook, we have a single task.  The task is that no matter the user, we need to provide a list of the recommendations based on simply the most popular items.

For this task, we will consider what is "most popular" based on the following criteria:

* A movie with the highest average rating is considered best
* With ties, movies that have more ratings are better
* A movie must have a minimum of 5 ratings to be considered among the best movies
* If movies are tied in their average rating and number of ratings, the ranking is determined by the movie that is the most recent rating

With these criteria, the goal for this notebook is to take a **user_id** and provide back the **n_top** recommendations.  Use the function below as the scaffolding that will be used for all the future recommendations as well.

### Note:
Problem in previous function will appear here, as due to filtering, many movie_id will be removed, so the checking will raise error.

So to solve this, Merging is needed between the Two dataframes, then get the movies_id after joining, not before it. 

````python
def popular_recommendations(user_id, n_top, years = None, genres = None):
    '''
    INPUT:
    user_id - the user_id of the individual you are making recommendations for
    n_top - an integer of the number recommendations you want back
    years - list of strings of selected years to filter movies
    genres - list of strings of selected genres to filter movies
    OUTPUT:
    top_movies - a list of the n_top recommended movies by movie title in order best to worst
    '''
    # A movie with the highest average rating is considered best
    # If movies are tied in their average rating and number of ratings, the ranking is determined by the movie that is the most recent rating
    #filtered_df =reviews.groupby('movie_id', as_index=False).agg(rate_mean =('rating','mean'),rate_counts = ('movie_id','count'),rate_recent=('date','max'))
    filtered_df =reviews.groupby('movie_id').agg(rate_avg =('rating','mean'),rate_counts = ('movie_id','count'),rate_recent=('date','max'))
    # merge with the movies dataset
    movies_records = movies.join(filtered_df, on = 'movie_id')
    
    # sort by top avg rating and number of ratings
    movies_ranked = movies_records.sort_values(['rate_avg', 'rate_counts', 'rate_recent'], ascending=False)

    # for edge cases - subset the movie list to those with only 5 or more reviews
    movies_ranked = movies_ranked[movies_ranked.rate_counts >= 5]
    #display(movies_ranked)
    
    # No filter applied
    if (years == None or len(years)==0) and (genres == None or len(genres)==0):
        movies_selected = movies_ranked
    # Filter applied    
    else:
        year_selected, genres_selected = [], []
        if years != None :
            # List of Boolean, True if movie is in selected year, else False
            year_selected = [True if str(year) in years else False for year in movies_ranked.date]

        if genres != None:
            # List of boolean lists, sum over all rows, So 0 means False, otherwise True 
            genres_selected = np.sum([movies_ranked[col]==1 for col in genres],axis=0)
    
        # Logical Or between years , genres
        movies_selected = movies_ranked[np.logical_or(year_selected,genres_selected)]

    if n_top > len(movies_selected.index):
        n_top = len(movies_selected.index)

    top_movies = list(movies_selected.movie[:n_top])
    
    return top_movies # a list of the n_top movies as recommended
````
